###imports section

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

###create Data Frames with Pandas

```python
R1400P1 = pd.read_excel("R1400.xlsx", sheet_name='R1400P1')
R1401P1 = pd.read_excel("R1401.xlsx", sheet_name='R1401P1')
R98P1 = pd.read_excel("R98.xlsx", sheet_name='R98P1')
R99P1 = pd.read_excel("R99.xlsx", sheet_name='R99P1')
U1400P1 = pd.read_excel("U1400.xlsx", sheet_name='U1400P1')
U98P1 = pd.read_excel("U98.xlsx", sheet_name='U98P1')
U99P1 = pd.read_excel("U99.xlsx", sheet_name='U99P1')
```

##Plots

###نمدار تحصیل


```python
fig, ax = plt.subplots(7, 1, figsize=(10, 30))
dataframes = [R1400P1, R1401P1, R98P1, R99P1, U1400P1, U98P1, U99P1]
labels = ["R1400", "R1401", "R98", "R99", "U1400", "U98", "U99"]

for i, df in enumerate(dataframes):
    degree_counts = df['degree'].value_counts()
    ax[i].bar(degree_counts.index, degree_counts.values, color="purple")
    ax[i].set_title(labels[i])
    ax[i].set_xlabel('Degree')
    ax[i].set_ylabel('Count')

fig.tight_layout()

plt.show()
```

###نمودار سواد

```python
for df in [R1400P1, R1401P1, R98P1, R99P1, U1400P1, U98P1, U99P1]:
    df.replace({'literate': 1, 'illiterate': 0}, inplace=True)

fig, ax = plt.subplots(7, 1, figsize=(10, 30))
dataframes = [R1400P1, R1401P1, R98P1, R99P1, U1400P1, U98P1, U99P1]
labels = ["R1400", "R1401", "R98", "R99", "U1400", "U98", "U99"]

for i, df in enumerate(dataframes):

    literate = df["literacy"].sum()
    illiterate = len(df) - literate

    ax[i].bar(["literate", "illiterate"], [literate, illiterate], color="purple")
    ax[i].set_title(labels[i])
    ax[i].set_xlabel('Literacy')
    ax[i].set_ylabel('Count')

fig.tight_layout()

plt.show()
```

###Grouped by

```python
dataframes = [R1400P1, R1401P1, R98P1, R99P1, U1400P1, U98P1, U99P1]

for i, df in enumerate(dataframes):
  df['degree'] = df['degree'].map({'Elemantry':1, 'Secondary':2, 'HighSchool':3, 'Diploma':4, 'College':5, 'Bachelor':6, 'Master':7, 'PhD':8, 'Other' :9})
  df.dropna(inplace=True)

fig, ax = plt.subplots(7, 1, figsize=(10, 30))
labels = ["R1400", "R1401", "R98", "R99", "U1400", "U98", "U99"]

for i, df in enumerate(dataframes):
  df_grouped = df.groupby('Address').agg({'degree': 'max'})
  group_counts_sorted = df_grouped['degree'].value_counts().sort_index()
  ax[i].bar(group_counts_sorted.index, group_counts_sorted.values, color="purple")
  ax[i].set_title(labels[i])
  ax[i].set_xlabel('Age')
  ax[i].set_ylabel('Count')

fig.tight_layout()
plt.show()
```
